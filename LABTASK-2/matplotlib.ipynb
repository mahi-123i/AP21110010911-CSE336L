```python
#Write a Python program to draw a line using given axis values taken from a text file, with suitable label in the x axis, y axis and a title.
import matplotlib.pyplot as plt

file_name = "mahesh.txt" 
with open(file_name, 'r') as file:
    lines = file.readlines()
    data = [list(map(float, line.strip().split())) for line in lines]
x = [item[0] for item in data]
y = [item[1] for item in data]
plt.plot(x, y, label='Line Plot', color='r')
plt.show()
```

```python
#Write a Python program to plot two or more lines with legends, different widths and colors.
import matplotlib.pyplot as plt
x1 = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
x2 = [1, 2, 3, 4, 5]
y2 = [1, 3, 5, 7, 9]
plt.plot(x1, y1, label='Line 1', color='blue', linewidth=2)
plt.plot(x2, y2, label='Line 2', color='red', linewidth=3)
plt.xlabel('X Axis Label')
plt.ylabel('Y Axis Label')
plt.show()
```

```python
#Write a Python program to plot quantities which have an x and y position.
import matplotlib.pyplot as plt
x= [1, 2, 3, 4, 5]
y= [2, 4, 6, 8, 10]
plt.scatter(x,y, label='Quantities', color='green', marker='o')
plt.xlabel('X Axis Label')
plt.ylabel('Y Axis Label')
plt.title('Scatter Plot of Quantities')
plt.show()
```

```python
#Create plot using subplots(), scatter, and some other methods on the Axes object:
import matplotlib.pyplot as plt
import numpy as np
x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50) * 100
fig, ax = plt.subplots()
scatter = ax.scatter(x, y, s=z, alpha=0.7, color='green', edgecolors='black')
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_title('Scatter Plot with Varying Marker Size')
cbar = plt.colorbar(scatter)
cbar.set_label('Bubble Size')
plt.show()
```

```python
#Generate random data for x and y and create a scatter plot using Matplotlib.
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(42)
num_points = 50
x = np.random.rand(num_points)
y = np.random.rand(num_points)
plt.scatter(x, y, color='blue', marker='o', alpha=0.5)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.show()
```

```python
#Create a bar chart to compare the sales of different products. You can use fictional data or come up with your own dataset.
import matplotlib.pyplot as plt
products = ['Product A', 'Product B']
sales = [100, 200]
plt.bar(products, sales, color='r')
plt.xlabel('Products')
plt.ylabel('Sales (units)')
plt.title('Product Sales Comparison')
plt.show()
```

```python
#Generate a random dataset and create a histogram using Matplotlib to visualize the distribution of
the data.
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(42)
random_data = np.random.normal(loc=50, scale=20, size=100)
plt.hist(random_data, bins=30, color='y', edgecolor='black')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram of Random Data')
plt.show()
```

```python
#Create a pie chart to represent the distribution of expenses in a budget. Use fictional data or create your own.
import matplotlib.pyplot as plt
categories = ['Housing', 'Utilities', 'Food', 'Transportation', 'Entertainment']
expenses = [1000, 300, 500, 250, 200]
plt.pie(expenses, labels=categories, autopct='%1.1f%%', 
        startangle=90, colors=['skyblue', 'lightcoral', 'lightgreen', 'lightsalmon', 'lightblue'])
plt.show()
```
